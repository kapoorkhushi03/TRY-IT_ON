```python
import matplotlib.pyplot as plt
from PIL import Image, ImageOps

# Load the image (replace 'path_to_your_image.jpg' with your image file)
image_path = "/content/output_mask.jpg"  # Change this to the path of your image
image = Image.open(image_path).convert("L")  # Convert to grayscale

# Invert the colors
inverted_image = ImageOps.invert(image)

# Display the result
plt.subplot(1, 2, 1)
plt.imshow(image, cmap="gray")
plt.axis("off")
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(inverted_image, cmap="gray")
plt.axis("off")
plt.title("Inverted Image")

plt.show()

# Save the result if needed
inverted_image.save("/content/inverted_image.jpg")  # Change the path if necessary
```
